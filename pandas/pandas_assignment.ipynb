# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should 
calculate and print the sum of the first three values, which is 60

```python
# Create a sample DataFrame
df2 = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Calculate the sum of the first three values in the 'Values' column
sum = 0
for index, row in df2.iterrows():
    if index < 3:
        sum += row['Values']

# Print the sum to the console
print(sum)
```

# Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the 
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your 
function should extract the username from each email address and store it in the new 'Username' 
column.

```python
def extract_username(df):
    df['username'] = df['Email'].apply(lambda x :x.split('@')[0])
    return df
```

```python
df5 = pd.DataFrame({'Email' : ['john.doe@example.com', 'jane.doe@example.com']})
df5 = extract_username(df5)
df5
```

# Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.

For example, if df contains the following values:

   A   B   C

0  3   5   1

1  8   2   7

2  6   9   4

3  2   3   5

4  9   1   2

Your function should select the following rows:   A   B   C

1  8   2   7

4  9   1   2

The function should return a new DataFrame that contains only the selected rows

```python
def select_rows(df):
    selected_rows = df[(df['A']>5) & (df["B"]<10)]
    return selected_rows
```

```python
df6 = pd.DataFrame({'A':[3,8,6,2,9],'B':[5,2,9,3,1],'C':[1,7,4,5,2]})
df6_new = select_rows(df6)
print(df6_new)
```

# Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
def cal_mean_med_std(df):
    df['Values'] = pd.to_numeric(df['Values'], errors='coerce')
    mean = df['Values'].mean()
    median = df['Values'].median()
    std = df['Values'].std()
    
    return mean , median ,std
```

```python
df7 = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
cal_mean_med_std(df7) 
```

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
df8 = pd.DataFrame({'Sales':[10,20,30,40,50,60,70,80,90],'Date':["2023-01-01","2023-01-02","2023-01-03","2023-01-04","2023-01-05","2023-01-06","2023-01-07","2023-01-08","2023-01-09"],
})
```

```python
df8
```

```python
def create_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df
```

```python
new_df = create_moving_average(df8).copy()
new_df
```

# Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a newcolumn 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:

         Date

0  2023-01-01

1  2023-01-02

2  2023-01-03

3  2023-01-04

4  2023-01-05

Your function should create the following DataFrame:


         Date    Weekday

0  2023-01-01    Sunday

1  2023-01-02     Monday

2  2023-01-03    Tuesday

3  2023-01-04    Wednesday

4  2023-01-05    Thursday

The function should return the modified DataFrame.

```python
df9 = pd.DataFrame({'Date':["2023-01-01","2023-01-02","2023-01-03","2023-01-04","2023-01-05","2023-01-06","2023-01-07","2023-01-08","2023-01-09"]})
```

```python
def weekdays(df):
    df["weekday"] = pd.to_datetime(df["Date"]).dt.day_name()
    return df
```

# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python

import pandas as pd
import numpy as np

# Define the start and end dates for the sales data
start_date = '2023-01-01'
end_date = '2023-12-31'

# Create a DatetimeIndex for the entire date range
date_index = pd.date_range(start=start_date, end=end_date, freq='D')

# Generate random sales data for each day in the date range
sales_data = np.random.randint(low=1000, high=10000, size=(len(date_index)))

# Create a Pandas DataFrame with the date index and sales data
df10 = pd.DataFrame({'Date': date_index, 'Sales': sales_data})

# Select only the rows that fall within the date range for January 2023
january_sales_data = df10[(df10['Date'] >= '2023-01-01') & (df10['Date'] <= '2023-01-31')]

# Print the January sales data
print(january_sales_data)
```

```python
import pandas as pd

def select_january_data(df):
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    january_data = df.loc[mask]
    return january_data
```
